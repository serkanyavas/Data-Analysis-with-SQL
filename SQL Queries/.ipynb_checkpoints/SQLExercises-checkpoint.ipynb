# SQL Queries for SuperHeros Dataset
* Data URL::https://www.kaggle.com/claudiodavi/superhero-set

* Created by Serkan YavaÅŸ

```python
import sqlite3 as sql
import pandas as pd
conn=sql.connect("Exercise.db")
df = pd.read_excel('heroes_information.xlsx')
df.to_sql("heroes_information",conn,if_exists='replace')
cur = conn.cursor()
```

#### Problem 1: Hero nums for every alignment

```python
cur.execute("SELECT Alignment, COUNT(*) FROM heroes_information GROUP BY Alignment ORDER BY COUNT(*) DESC ")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 2: Average superheros weight by Publisher

```python
cur.execute("SELECT Publisher, AVG(Weight) as ort FROM heroes_information GROUP BY Publisher HAVING ort>0")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 3: Most Common Hair Colors

```python
cur.execute("SELECT Hairc, COUNT(*) FROM heroes_information GROUP BY Hairc ORDER BY COUNT(*) DESC ")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 4: The Tallest Superhero


```python
cur.execute("SELECT name ,Height FROM heroes_information WHERE Height = (SELECT max(Height)FROM heroes_information)")
rows = cur.fetchall()
for row in rows:
        print(row)
        
```

#### Problem 5: Most Common Eyecolors

```python
cur.execute("SELECT Eyec, COUNT(*) FROM heroes_information GROUP BY Eyec ORDER BY COUNT(*) DESC ")
rows = cur.fetchall()
for row in rows:
        print(row)
        
```

#### Problem 6: The Heaviest Superhero

```python
cur.execute("SELECT name ,Weight FROM heroes_information WHERE Weight = (SELECT max(Weight)FROM heroes_information)")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 7: Bad Superheros

```python
cur.execute("select name, alignment from heroes_information where Alignment='bad'")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 8: Most Common Skin Colors

```python
cur.execute("SELECT Skinc, COUNT(*) FROM heroes_information GROUP BY Skinc ORDER BY COUNT(*) DESC ")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 9: Average superheros height by Gender

```python
cur.execute("SELECT Gender, AVG(Height) as ort FROM heroes_information GROUP BY Gender HAVING ort>50")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 10: Names of Superheros created by Marvel

```python
cur.execute("SELECT name, Publisher FROM heroes_information WHERE Publisher='Marvel Comics'")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 11: Most Common Races

```python
cur.execute("SELECT race, COUNT(*) FROM heroes_information GROUP BY race ORDER BY COUNT(*) DESC ")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 12: Heights in between 10-100 

```python
cur.execute("SELECT Height,name FROM heroes_information WHERE Height BETWEEN \'10\' AND \'100\'")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 13: Height Ranking

```python
cur.execute("SELECT name, Height FROM heroes_information ORDER BY Height desc")
rows = cur.fetchall()
for row in rows:
        print(row)
```

#### Problem 14: Nums of Created Superheros by Publisher

```python
cur.execute("SELECT Publisher, COUNT(*) FROM heroes_information GROUP BY Publisher ORDER BY COUNT(*) DESC ")
rows = cur.fetchall()
for row in rows:
        print(row)
```
